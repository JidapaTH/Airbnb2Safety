## Bag of words (Austin)

```python
import importlib.util
spec = importlib.util.spec_from_file_location("nltk_pack", "../../toolkit/nltk_pack.py")
pack = importlib.util.module_from_spec(spec)
spec.loader.exec_module(pack)
import pickle
import numpy as np
import pandas as pd
filePath = "./"
```

### Load the review data

```python

# listing = pd.read_csv(filePath)

listing = pickle.load( open( filePath +"review_by_zipcode_AU.p", "rb" ) )
```

```python
cleanlist = listing.dropna() # clean NA
len(cleanlist)
```

### Run text cleansing & count vectorizor function

```python
# create feature
feature =  pack.getfeature2(cleanlist['review'], ngram = 3)
```

```python
feature.index = cleanlist.index.values
feature.head()
```

### Find the proportion of words in each area (Count/Total reviews)

```python

len(cleanlist['ct'].replace(0, 1))
feature2 = feature.div(cleanlist['ct'].replace(0, 1), axis=0) # divide count by total amount of reviews
wordbag =  pd.concat([cleanlist, feature2], axis=1)
wordbag2 = wordbag.drop(['review','ct'], axis=1).dropna() # drop unnecessary columns
wordbag2.head()
```

```python
# wordbag.dropna().to_csv("listingwordbag.csv")

# dump file
with open(filePath + 'listingwordbag2_au.p', 'wb') as handle:
    pickle.dump(wordbag2.dropna(), handle, protocol=pickle.HIGHEST_PROTOCOL)
```

### Get important words according to LA data

```python
from sklearn.preprocessing import normalize

# load file
wordbag2 = pickle.load( open( filePath + 'listingwordbag2_au.p', "rb" ) )

# normalize
wordbag3 = pd.DataFrame(normalize(wordbag2 , axis=0), columns = wordbag2.columns, index = wordbag2.index.values)

wfilepath = "../../"
wordrank = pickle.load( open( wfilepath +"wordrank.pickle", "rb" ) )
wordrank[0:20]
```

```python
# if there are some important words from LA that don't appear in Austin data, they will be filled with 0.0
wr = pd.DataFrame(wordrank)

for w in wr[wr[0].isin(wordbag3.columns) == False][0]:
    print(w)
    wordbag3[w] = 0.0
```

```python
wordbag3[wordrank[0:20]]
```

```python
# get important word from LA data

column = wordbag3.columns[wordbag3.columns.isin(wordrank)]
data = wordbag3[column]
```

### Sentimental score

```python
# use textblob library to analyzed the sentimental score

from textblob import TextBlob

senti_score = [(TextBlob(w).sentiment.polarity) if (TextBlob(w).sentiment.polarity) > 0 else 10*(TextBlob(w).sentiment.polarity) for w in data.iloc[0].index]
data['sentimental'] = normalize(pd.DataFrame([np.sum(row*senti_score) for row in np.array(data)]))
```

```python
# Calculate average per reviews of each area as a feature

data['wordperrev'] = data.drop('sentimental', axis=1).sum(axis=1)
data.head()
```

```python
# save

path = "./wordbag_AU.pickle"
pickle.dump( data, open(  path, "wb") )
# use the same wordrank as LA
```
