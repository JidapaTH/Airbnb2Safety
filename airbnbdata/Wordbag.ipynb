## Bag of words

```python
import importlib.util
spec = importlib.util.spec_from_file_location("nltk_pack", "../toolkit/nltk_pack.py")
pack = importlib.util.module_from_spec(spec)
spec.loader.exec_module(pack)
import pickle
import numpy as np
import pandas as pd
```

### Load the review data

```python
# filePath = "../../"
# listing = pd.read_csv(filePath)

# listing = pickle.load( open( filePath +"review_by_zipcode.p", "rb" ) )


filePath = "./"
# listing = pd.read_csv(filePath)

listing = pickle.load( open( filePath +"balanced_review_by_zipcode.p", "rb" ) )
```

```python
cleanlist = listing.dropna() # clean NA
len(cleanlist)
```

### Run text cleansing & count vectorizor function

```python
# create feature
feature =  pack.getfeature2(cleanlist['review'], ngram = 2)
```

```python
feature.index = cleanlist.index.values
feature.head()
```

### Find the rate of the appearance of words (Count/Total reviews)

```python
# find proportion
len(cleanlist['ct'].replace(0, 1))
feature2 = feature.div(cleanlist['ct'].replace(0, 1), axis=0) # divide count by total amount of reviews
wordbag =  pd.concat([cleanlist, feature2], axis=1)
# wordbag.div(wordbag['ct'].replace(0, 1), axis=0, fill_value=None)

wordbag.head()
```

```python
# normalize and save the result

filePath = "../../"
wordbag = pickle.load( open( filePath + 'listingwordbag2.p', "rb" ) )
#normalize
from sklearn.preprocessing import normalize
wordbag = wordbag.drop('review', axis=1)
wordbag = pd.DataFrame(normalize(wordbag , axis=0), columns = wordbag.columns, index = wordbag.index.values)
wordbag.head()
```

### Merge with crime score

```python
#  load crime score
filePath = "../Crime/safety_score_LA.csv"
crime = pd.read_csv(filePath)

crime.head()
```

```python
# merge with crime score by using zipcode

wordbag = wordbag.dropna()
wordbag['zipcode'] = wordbag.index.values
data = pd.merge(wordbag, crime[['ZIP','level2']], left_on='zipcode', right_on='ZIP', how='left', sort=False)
data = data.drop(['ct','ZIP'], axis=1).dropna()
```

```python
# set index and check data
data = data.set_index('zipcode')
print(data.head())
np.unique(data['level2'])
```

### For the important word by the random forest classifier

```python
from sklearn.model_selection import train_test_split


# split data into train and test set
X_train, X_test, y_train, y_test = train_test_split(\
    data.drop('level2', axis = 1), data['level2'], random_state=0, test_size=.2)
```

```python
# train the model
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 50)
forest = forest.fit(X_train.astype(int), y_train.astype(int)) 
```

```python
# see the accuracy

from sklearn import metrics 
train_predictions = forest.predict(X_train.astype(int))
valid_predictions = forest.predict(X_test.astype(int))
print(metrics.accuracy_score(y_train.astype(int),train_predictions))
print(metrics.accuracy_score(y_test.astype(int),valid_predictions))
```

### Get important words from the model result

```python
importances = forest.feature_importances_

# sort importance scores
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")
top_30 = indices[:30]

# Get top ten features
print(X_train.columns[top_30])
```

### Calculate sentimental score from review of each zip

```python
from textblob import TextBlob

# get the sentimental score of each word
# we have positive reviews 10 times more than negative so give weight *5
senti_score = [(TextBlob(w).sentiment.polarity) if (TextBlob(w).sentiment.polarity) > 0 else 5*(TextBlob(w).sentiment.polarity) for w in data.iloc[0].index]
```

```python
# Area sentimental score=all words(Sentimental score * proportion of each word in each area)
# multiply score and the proportion of words then sum together
data['sentimental'] = [np.sum(row*senti_score) for row in np.array(data)]


# Calculate average words per each review in each area as another feature
data['wordperrev'] = data.drop('sentimental', axis=1).sum(axis=1)
```

### Neighborhood overview

```python
filePath = "../../"
neighbor = pd.read_csv(filePath+ "listings.csv")  # use cleaned listing to join with review data
```

```python
neigh = neighbor[['zipcode','neighborhood_overview']].dropna() # clean NA
neigh.head()
```

```python
# use counter to count the word from neighborhood overview

from collections import Counter
import re
neigh['count'] = [sum(Counter(re.findall(r"[\w']+",w)).values()) for w in neigh['neighborhood_overview']]
```

Find the keywords from word embedding

```python
# Count the safety keyword in Neighborhood description
# we derived these keyword by skimming through some reviews

safetyKeyWd = ["good", "safe","peaceful","night", "walk", "unsafe", "bad", "dangerous"] 

count = [Counter(re.findall(r"[\w']+",w)) for w in neigh['neighborhood_overview']]
for w in safetyKeyWd:
    neigh[w] = [wc[w] for wc in count]
neigh.head()
```

```python
# Group and average by zipcode
neighfeature = neigh.groupby('zipcode').mean()
    path = "../neighborhoodfeature.pickle"
pickle.dump(neighfeature, open(  path, "wb") )
```
