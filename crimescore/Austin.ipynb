```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

file_path='2016_Annual_Crime_Data.csv'
data=pd.read_csv(file_path, sep=',', header=0, delimiter=',',encoding='iso-8859-1')
```

```python
data
```

```python
 np.unique(data['Highest NIBRS/UCR Offense Description'])
```

```python
category=pd.read_csv('category_Austin.csv', sep=',', header=0, delimiter=',')
category.head()
```

```python
crime = data.dropna().drop(['GO Primary Key','Council District','GO Highest Offense Desc','GO Report Date','GO Location','Clearance Status'], axis=1)
```

```python
crime = crime.dropna().drop(['Clearance Date','Clearance Date','GO District','GO Census Tract','GO X Coordinate','GO Y Coordinate'], axis=1)
```

```python
crime
```

```python
crime.columns = ['CATEGORY','ZIP']
```

```python
crime2 = pd.merge(crime, category, left_on='CATEGORY', right_on='CATEGORY', how='left', sort=False)
```

```python
crime2 = crime2.dropna().drop(['Unnamed: 0','crime'],axis =1)
```

```python
freq = crime2.groupby(by=['ZIP']).sum()[['score']]
```

```python
freq=freq.reset_index()
freq.head()
```

```python
zip_pop =pd.read_csv('pop-by-zip-code.csv', sep=',', header=0, delimiter=',')
zip_pop = zip_pop[['zip_code','y-2016']]
```

```python
zip_pop['zip_code']=zip_pop['zip_code']*1.0
```

```python
d = zip_pop.set_index('zip_code')['y-2016'].to_dict()
```

```python
freq['pop'] = freq['ZIP'].map(d)
```

```python
freq 
freq['pop'].fillna(1, inplace=True)
```

```python
freq['score/population'] = freq['score']/freq['pop']
freq.head()
```

```python
freq['log'] = np.log(freq['score/population'])
```

```python
import pickle
filePath = "./"
# listing = pd.read_csv(filePath)

bins = pickle.load( open( filePath +"bins.pickle", "rb" ) )
print(bins)
freq['level2'] = pd.qcut(freq['log'], 3,labels=[1,2,3],duplicates='drop')
```
