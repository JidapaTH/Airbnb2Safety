```python
from shapely.geometry import Point
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import geopandas as gpd
%matplotlib inline
```

```python
file_path='sf_crime.csv'
data=pd.read_csv(file_path, sep=',', header=0, delimiter=',',encoding='iso-8859-1')
```

```python
data['geometry'] = [Point((i[0], i[1])) for i in np.array(data[['X','Y']])]
```

```python

zip = gpd.read_file('sf_zip.geojson')
print(zip.head())
```

```python
np.unique(list(data['PdDistrict']))
```

```python
gdf = gpd.GeoDataFrame(data, geometry='geometry').dropna()
```

```python
gdf.head()
```

```python
join = gpd.sjoin(zip, gdf, how='inner', op='contains', lsuffix='left', rsuffix='right')
        
```

```python
join
```

```python
join.columns
```

```python
sf_crime_withzip = join[['area', 'zip', 'state', 'po_name', 'length', 'index_right',
       'IncidntNum', 'Category', 'Descript', 'DayOfWeek', 'Date', 'Time',
       'PdDistrict', 'Resolution', 'Address', 'X', 'Y']]

sf_crime_withzip.to_csv('sf_crime_withzip.csv', sep=',', encoding='utf-8')
```

```python
file_path='sf_crime_withzip.csv'
data=pd.read_csv(file_path, sep=',', header=0, delimiter=',',encoding='iso-8859-1')
```

```python

data['Datetime'] = pd.to_datetime(data['Date'].astype(str) +' ' + data['Time'], format='%m/%d/%Y %H:%M')
data.head()
```

```python
data['year'] = [i.year for i in data['Datetime']]
```

```python
len(data[data['year'] == 2017] )
```

```python
data[data['year'] == 2017].to_csv('sf_crime_withzip.csv', sep=',', encoding='utf-8')
```
